```python
import pandas as pd

in_file = '/content/drive/My Drive/GEE_Downloads/big_fc.csv'
out_file = '/content/drive/My Drive/GEE_Downloads/rcp_gcm_list.csv'

gcms = ['ACCESS1-0', 'bcc-csm1-1', 'bcc-csm1-1-m', 'BNU-ESM', 'CanESM2', 'CCSM4', 'CESM1-BGC', 'CESM1-CAM5', 'CMCC-CM', 'CNRM-CM5', 'CSIRO-Mk3-6-0', 'FGOALS-g2', 'FIO-ESM', 'GFDL-CM3', 'GFDL-ESM2G', 'GFDL-ESM2M', 'GISS-E2-H-CC', 'GISS-E2-R', 'GISS-E2-R-CC', 'HadGEM2-AO', 'HadGEM2-CC', 'HadGEM2-ES', 'inmcm4', 'IPSL-CM5A-LR', 'IPSL-CM5A-MR', 'IPSL-CM5B-LR', 'MIROC5', 'MIROC-ESM', 'MIROC-ESM-CHEM', 'MPI-ESM-LR', 'MPI-ESM-MR', 'MRI-CGCM3', 'NorESM1-M']
scens = ['rcp26', 'rcp45', 'rcp60', 'rcp85']

with open(in_file) as f:
  lines = f.readlines()

full_list = []
for l in lines[1:]:
  model = l.split('"[')[1].split(',')[0]
  scen = l.split(']"')[0].split(',')[-1].lstrip()
  if scen != 'historical':
    full_list.append((model, scen))


set_list = list(set(full_list))
print(set_list)

print(len(set_list))



df = pd.DataFrame(columns=scens, index=gcms)

for tpl in set_list:
  model = tpl[0]
  gcm = tpl[1]
  df.at[model, gcm] = 1

print(df)

rcp26_gcms = df[df['rcp26'] > 0].index
rcp45_gcms = df[df['rcp45'] > 0].index
rcp60_gcms = df[df['rcp60'] > 0].index
rcp85_gcms = df[df['rcp85'] > 0].index

rcp_list = [rcp26_gcms, rcp45_gcms, rcp60_gcms, rcp85_gcms]

with open(out_file, 'w') as fo:
  fo.write('rcp,gcms\n')
  for i, rcp in enumerate(['rcp26', 'rcp45', 'rcp60', 'rcp85']):
    fo.write(rcp + ',' + ','.join(rcp_list[i]) + '\n')



```
